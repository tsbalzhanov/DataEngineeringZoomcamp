```python
import pandas as pd
import sqlalchemy
```

```python
trip_df = pd.read_csv(
    'data/green_tripdata_2019-09.csv',
    parse_dates=['lpep_pickup_datetime', 'lpep_dropoff_datetime'],
    low_memory=False
)
print(len(trip_df))
trip_df.head()
```

```python
sql_url = 'postgresql://admin:admin@localhost/nytaxi'
sql_engine = sqlalchemy.create_engine(sql_url)
```

```python
%%time
trips_table_name = 'trips'

with sql_engine.connect() as conn:
    trip_df.to_sql(trips_table_name, conn, if_exists='replace', chunksize=50000)
```

```python
zone_df = pd.read_csv('data/taxi_zone_lookup.csv')
print(len(zone_df))
zone_df.head()
```

```python
%%time
zone_table_name = 'zones'

with sql_engine.connect() as conn:
    zone_df.to_sql(zone_table_name, conn, if_exists='replace')
```
